#Data Processing

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df = pd.read_csv('https://raw.githubusercontent.com/DSomair/Masar_DS/main/Datasets/Week2/Masar_clean.csv')
```

```python
df
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

```python
columns_to_normalize = ['AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']

# Normalization using StandardScaler
scaler = StandardScaler()
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
```

```python
df
```

```python
# Scaling to a specific range [0, 1] using MinMaxScaler
min_max_scaler = MinMaxScaler()
df[columns_to_normalize] = min_max_scaler.fit_transform(df[columns_to_normalize])
```

```python
df.head()
```

```python
# Replace 'F' with 0 and 'M' with 1 in the 'Gender' column
df['Gender'] = df['Gender'].replace({'F': 0, 'M': 1})
```

```python
df.head()
```

```python
# Replace 'N' with 0, 'P' with 1, and 'Y' with 2 in the 'CLASS' column
df['CLASS'] = df['CLASS'].replace({'N': 0, 'P': 1, 'Y': 2})
```

```python
# Creating new variable

import pandas as pd

# Create a sample dataset with weight (in kg) and height (in meters)
data = {'Weight': [65, 70, 80, 75, 68],
        'Height': [1.75, 1.80, 1.65, 1.70, 1.72]}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Compute BMI and add a new column 'BMI' to the DataFrame
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

# Display the DataFrame with BMI
print("\nDataFrame with BMI:")
print(df)
```

#Data aggregation

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame for sales transactions
data = {
    'Date': pd.date_range(start='2023-01-01', periods=20),
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A', 'B'],
    'Amount': np.random.randint(100, 1000, size=20),
    'Region': ['North', 'South', 'East', 'West'] * 5
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Group by 'Product' and 'Region', calculate the total amount and average amount for each group
grouped_data = df.groupby(['Product', 'Region']).agg({'Amount': ['sum', 'mean']}).reset_index()

# Display the grouped data
print("\nGrouped Data by Product and Region (sum and mean of Amount):")
print(grouped_data)

# Filter rows where the total amount is greater than 1500
filtered_data = grouped_data[grouped_data[('Amount', 'sum')] > 1500]

# Display the filtered data
print("\nFiltered Data (Total Amount > 1500):")
print(filtered_data)

# Create a pivot table to summarize data by 'Product' and 'Region', showing total amount and average amount
pivot_table = pd.pivot_table(df, values='Amount', index=['Product'], columns=['Region'], aggfunc={'Amount': ['sum', 'mean']}, fill_value=0)

# Display the pivot table
print("\nPivot Table:")
print(pivot_table)
```

```python
table1 = pd.DataFrame({"P_ID" : (1,2,3,4,5,6,7,8),
                     "gender" : ("male", "male", "female","female",
                                "female", "male", "female", "male"),
                     "height" : (71,73,64,64,66,69,62,72),
                     "weight" : (175,225,130,125,165,160,115,250)})

table1
```

```python
table2 = pd.DataFrame({"P_ID" : (1, 2, 4, 5, 7, 8, 9, 10),
                     "sex" : ("male", "male", "female","female",
                            "female", "male", "male", "female"),
                     "visits" : (1,2,4,12,2,2,1,1),
                     "checkup" : (1,1,1,1,1,1,0,0),
                     "follow_up" : (0,0,1,2,0,0,0,0),
                     "illness" : (0,0,2,7,1,1,0,0),
                     "surgery" : (0,0,0,2,0,0,0,0),
                     "ER" : ( 0,1,0,0,0,0,1,1) } )

table2
```

```python
combined1 = pd.merge(table1,       # First table
                    table2,        # Second table
                    how="inner",   # Merge method
                    on="P_ID")     # Column(s) to join on

combined1
```

```python
# A left join keeps all key values in the first(left) data frame

left_join = pd.merge(table1,       # First table
                    table2,        # Second table
                    how="left",   # Merge method
                    on="P_ID")     # Column(s) to join on

left_join
```

```python
# A right join keeps all key values in the second(right) data frame

right_join = pd.merge(table1,       # First table
                    table2,        # Second table
                    how="right",   # Merge method
                    on="P_ID")     # Column(s) to join on

right_join
```
