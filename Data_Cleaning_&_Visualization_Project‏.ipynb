```python
columns_to_normalize = ['Open',	'High',	'Low',	'Close',	'Adj Close',	'Volume']

# Normalization using StandardScaler
scaler = StandardScaler()
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
```

```python
min_max_scaler = MinMaxScaler()
df[columns_to_normalize] = min_max_scaler.fit_transform(df[columns_to_normalize])
```

```python
import matplotlib.pyplot as plt

data = {
    'Open': [1980, 1970, 1980, 1975, 2060],
    'High': [2000, 1980, 1990, 2030, 2060],
    'Law': [1965, 1940, 1965, 1970, 1935],
    'Close': [1965, 1970, 1990, 2030, 2050]
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

plt.figure(figsize=(15, 6))
plt.bar(df['Open'], df['High'], label='High', color='blue', alpha=0.7)
plt.bar(df['Open'], df['Law'], label='Law', color='green', alpha=0.7)
plt.bar(df['Open'], df['Close'], label='Close', color='red', alpha=0.7)
plt.xlabel('Open')
plt.ylabel('Amount')
plt.title('High, Law, and Close Over Open')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(df['Open'], df['High'], marker='o', linestyle='-', color='blue')
plt.xlabel('Open')
plt.ylabel('High')
plt.title('High Over Open (Line Chart)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Law'], df['Close'], color='green', alpha=0.7)
plt.xlabel('Law')
plt.ylabel('Close')
plt.title('Law vs Close (Scatter Plot)')
plt.grid(True)
plt.show()
```
